```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# plotting box plot
# using the orient parameter for orientation of each numeric variable
sns.boxplot( data = df, orient="h")

# display
plt.show()


#Outliers Removal
def outliers (df , ft) :
    Q1= df[ft].quantile(0.25)
    Q3= df[ft].quantile(0.75)
    
    IQR = Q3-Q1
    
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    ls = df.index[(df[ft] < lower_bound )| (df[ft] > upper_bound )]
    
    return ls

def remove (df, ls ):
    ls = sorted(set(ls))
    df = df.drop(ls)
    return df
```

```python
index_list = []
for feature in ['SITE NUMBER','LICENSE CODE','LICENSE NUMBER','LATITUDE','LONGITUDE']:
    index_list.extend(outliers(df,feature))
    
df1 = remove(df,index_list)
print(df1.shape)
```

```python
#Taking Top 10 ten categories from categorical columnsand making them into other columns

max_categories = 10
cat_cols = ['LEGAL NAME','DOING BUSINESS AS NAME','STATE','LICENSE DESCRIPTION','APPLICATION TYPE']
# Extract the target variable from the DataFrame
target_col = 'LICENSE STATUS'
target = df1[target_col]

# loop through each categorical column
for col in cat_cols:
    # calculate the frequency of each category
    freq = df1[col].value_counts(normalize=True)
    
    # extract the top `max_categories` categories
    top_categories = freq.index[:max_categories]
    
    # create a new category for all other categories
    df1.loc[~df1[col].isin(top_categories), col] = 'other'
    
    # apply one-hot encoding to the column
    cat_transformed = pd.get_dummies(df1[col],drop_first=True, prefix=col)
    
    # concatenate the new one-hot encoded column to the transformed DataFrame
    df1 = pd.concat([df1, cat_transformed], axis=1)
    
    # drop the original categorical column
    df1.drop(columns=[col], inplace=True)

# Add the target variable back to the transformed DataFrame
df1[target_col] = target


# Perform any additional data preprocessing or modeling steps as needed

print(df1.shape)
print(df1.dtypes)

```

```python
# Convert all datetime columns to days from a date
datetime_cols = [col for col in df1.columns if df1[col].dtype == 'datetime64[ns]']
for col in datetime_cols:
    df1[f'{col}_days'] = (df1[col] - pd.Timestamp('1970-01-01')).dt.days
    
# Drop the original datetime columns
df1 = df1.drop(datetime_cols, axis=1)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)




```

```python
#Randaom forest classifier with Gridsearchcv

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, classification_report, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

# Define the parameter grid to search over
param_grid = {
    'n_estimators': [500],
    'max_depth': [6],
    'max_features': ['auto'],
    'bootstrap': [False]
}

# Create the GridSearchCV object
RFR = RandomForestClassifier(random_state=100)
grid_search = GridSearchCV(estimator=RFR, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit the GridSearchCV object to the data
grid_search.fit(X_train,y_train)

# Print the best parameters and the corresponding score
print("Best parameters found: ", grid_search.best_params_)
print("Best accuracy score: ", grid_search.best_score_)

# Use the best model to predict on the test set
best_RFR = grid_search.best_estimator_
Y_pred_tes = best_RFR.predict(X_test)

# Calculate and print the accuracy, precision, f1-score and recall
RFR_acc = accuracy_score(y_test, Y_pred_tes)
Precesion_RFR = precision_score(y_test,Y_pred_tes,average='macro')
f1score_RFR = f1_score(y_test,Y_pred_tes,average='macro')
recall_RFR = recall_score(y_test,Y_pred_tes,average='macro')

print("Test accuracy score: ", RFR_acc)
print("Precision score is: ", Precesion_RFR.round(2))
print("F1 score is: ", f1score_RFR.round(2))
print("Recall score is: ", recall_RFR.round(2))

# Print the classification report and confusion matrix
print(classification_report(y_test, Y_pred_tes))
sns.set_style("whitegrid", {'axes.grid' : False})
ConfusionMatrixDisplay.from_predictions(y_test, Y_pred_tes)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from bayes_opt import BayesianOptimization

# Define the XGBoost model
def xgb_cv(learning_rate, max_depth, gamma, subsample, colsample_bytree):
    model = XGBClassifier(
        learning_rate=learning_rate,
        max_depth=int(max_depth),
        gamma=gamma,
        subsample=subsample,
        colsample_bytree=colsample_bytree
    )
    return cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy').mean()

# Define the hyperparameter space to search
pbounds = {
    'learning_rate': (0.01, 0.3),
    'max_depth': (3, 10),
    'gamma': (0, 1),
    'subsample': (0.5, 1),
    'colsample_bytree': (0.5, 1)
}

# Perform the Bayesian optimization
optimizer = BayesianOptimization(
    f=xgb_cv,
    pbounds=pbounds,
    verbose=2,
    random_state=42
)
optimizer.maximize(init_points=5, n_iter=3)

# Print the best hyperparameters and score
print(optimizer.max)


from sklearn.metrics import classification_report

# Train the XGBoost model
xg = XGBClassifier()
xg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = xg.predict(X_test)

# Generate the classification report
report = classification_report(y_test, y_pred)
print(report)

Auc_XG_Baye = accuracy_score(y_test, y_pred)
Precesion_XG_Baye = precision_score(y_test,y_pred,average='macro')
f1score_XG_Baye = f1_score(y_test,y_pred,average='macro')
recall_XG_Baye = recall_score(y_test,y_pred,average='macro')
print("Precesion score is ",Precesion_XG_Baye.round(2))
print("F1 score is ",f1score_XG_Baye.round(2))
print(" recall ",recall_XG_Baye.round(2))

```
